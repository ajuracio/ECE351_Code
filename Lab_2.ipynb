Part 1

```python
#Task 1: libraries
import numpy as np
import matplotlib.pyplot as plt
 
#Task 2: func1 for y = cos(t)
def func1(t):
    y = np.zeros((len(t),1))

    for i in range(len(t)):
        y[i] = np.cos(t[i])
    return y
```

```python
#Task 2: graph y = cos(t)
steps = .1
t = np.arange(0,10+steps,steps)
y = func1(t)

myFigSize = (10,8)
plt.figure(figsize=myFigSize)
plt.subplot(2,1,1)
plt.plot(t,y)
plt.grid(True)
plt.xlabel('time')
plt.ylabel('y')
plt.title('y = cos(t)')
plt.show()
```

Part 2

$y(t)=r(t)-r(t-3)+5u(t-3)-2u(t-6)-2r(t-6)$

```python
#Task 1: define unit step and ramp functions
def u(t):
    if t < 0:
        return 0
    if t >= 0:
        return 1
    
def r(t):
    if t < 0: 
        return 0
    if t >= 0:
        return t   
```

```python
#Task 1: Implement 
def arrayconvert (t, func):
    y = np.zeros((len(t), 1))
    for i in range(len(t)):
        y[i] = func(t[i])
    return y
```

```python
#Task 1: Plot functions
t = np.arange(-5,10+steps,steps)
y = arrayconvert(t, u)

myFigSize = (10,8)
plt.figure(figsize=myFigSize)
plt.subplot(2,1,1)
plt.plot(t,y)
plt.grid(True)
plt.xlabel('time')
plt.ylabel('y')
plt.title('y = u(t)')
plt.show()
```

```python
t = np.arange(-5,10+steps,steps)
y = arrayconvert(t, r)

myFigSize = (10,8)
plt.figure(figsize=myFigSize)
plt.subplot(2,1,1)
plt.plot(t,y)
plt.grid(True)
plt.xlabel('time')
plt.ylabel('y')
plt.title('y = r(t)')
plt.show()
```

```python
def f(t): 
    y = np.zeros((len(t), 1))
    for i in range(len(t)):
        y[i] = r(t[i])-r(t[i]-3)+5*u(t[i]-3)-2*u(t[i]-6)-2*r(t[i]-6)
    return y

t = np.arange(-5,10+steps,steps)
y = f(t)

myFigSize = (10,8)
plt.figure(figsize=myFigSize)
plt.subplot(2,1,1)
plt.plot(t,y)
plt.grid(True)
plt.xlabel('time')
plt.ylabel('y')
plt.title('Plot 1')
plt.show()
```

```python
t = np.arange(-10,5+steps,steps)
y = f(-t)

myFigSize = (10,8)
plt.figure(figsize=myFigSize)
plt.subplot(2,1,1)
plt.plot(t,y)
plt.grid(True)
plt.xlabel('time')
plt.ylabel('y')
plt.title('Plot 1 with time reversal')
plt.show()
```
